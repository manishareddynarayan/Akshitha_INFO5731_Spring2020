```
import pycountry_convert as py
import pandas as pd
#Reading the 
read_file = pd.read_excel("HIV_input.xlsx")
input_var= {"AS":"Asia","AF":"Africa","NA":"North America","SA":"South America","EU":"Europe","OC":"Australia"}
x=[]
for country in read_file['Estimated HIV Prevalence% - (Ages 15-49)']:
    try:
        continent = input_var[py.country_alpha2_to_continent_code(py.country_name_to_country_alpha2(country, cn_name_format="default"))]
        x.append(continent)
    except:
        x.append('unknown continent')
read_file['Continent'] = x
read_file.to_csv("HIV_Output.csv")    
```

```
import csv
import pandas as pd
import numpy as np
import array as ar
import matplotlib.pyplot as plt
from pandas import DataFrame
import random 
import datascience as ds
#calculating mean for Asia from 2000 to 2011
read_file = pd.read_csv("HIV_Output.csv")
asia1 = read_file.loc[read_file['Continent'] == 'Asia']
asia2 = pd.concat([asia1['Estimated HIV Prevalence% - (Ages 15-49)'],asia1['Continent'],asia1.loc[:,"2000":"2011"].mean(axis=1)],axis=1)
asia2.columns = ['Region','Continent','Average']
Asia_highest_Avg=asia2[asia2['Average'] == asia2['Average'].max()]
Asia_lowest_Avg=asia2[asia2['Average'] == asia2['Average'].min()]
#calcuclating mean for Africa from 2000 to 2011
africa1 = read_file.loc[read_file['Continent'] == 'Africa']
africa2 = pd.concat([africa1['Estimated HIV Prevalence% - (Ages 15-49)'],africa1['Continent'],africa1.loc[:,"2000":"2011"].mean(axis=1)],axis=1)
africa2.columns = ['Region','Continent','Average']
Africa_highest_Avg=africa2[africa2['Average'] == africa2['Average'].max()]
Africa_lowest_Avg=africa2[africa2['Average'] == africa2['Average'].min()]
#calculating mean for North America from 2000 to 2011
NAmerica1 = read_file.loc[read_file['Continent'] == 'North America']
NAmerica2 = pd.concat([NAmerica1['Estimated HIV Prevalence% - (Ages 15-49)'],NAmerica1['Continent'],NAmerica1.loc[:,"2000":"2011"].mean(axis=1)],axis=1)
NAmerica2.columns = ['Region','Continent','Average']
North_America_highest_Avg=NAmerica2[NAmerica2['Average'] == NAmerica2['Average'].max()]
North_America_lowest_Avg=NAmerica2[NAmerica2['Average'] == NAmerica2['Average'].min()]
#calculating mean for South America from 2000 to 2011
SAmerica1 = read_file.loc[read_file['Continent'] == 'South America']
SAmerica2 = pd.concat([SAmerica1['Estimated HIV Prevalence% - (Ages 15-49)'],SAmerica1['Continent'],SAmerica1.loc[:,"2000":"2011"].mean(axis=1)],axis=1)
SAmerica2.columns = ['Region','Continent','Average']
South_America_highest_Avg=SAmerica2[SAmerica2['Average'] == SAmerica2['Average'].max()]
South_America_lowest_Avg=SAmerica2[SAmerica2['Average'] == SAmerica2['Average'].min()]
#calculating mean for Europe from 2000 to 2011
Europe1 = read_file.loc[read_file['Continent'] == 'Europe']
Europe2 = pd.concat([Europe1['Estimated HIV Prevalence% - (Ages 15-49)'],Europe1['Continent'],Europe1.loc[:,"2000":"2011"].mean(axis=1)],axis=1)
Europe2.columns = ['Region','Continent','Average']
Europe_highest_Avg=Europe2[Europe2['Average'] == Europe2['Average'].max()]
Europe_lowest_Avg=Europe2[Europe2['Average'] == Europe2['Average'].min()]
#Calculating mean for Australia from 2000 to 2011
Australia1 = read_file.loc[read_file['Continent'] == 'Australia']
Australia2 = pd.concat([Australia1['Estimated HIV Prevalence% - (Ages 15-49)'],Australia1['Continent'],Australia1.loc[:,"2000":"2011"].mean(axis=1)],axis=1)
Australia2.columns = ['Region','Continent','Average']
Australia_highest_Avg=Australia2[Australia2['Average'] == Australia2['Average'].max()]
Australia_lowest_Avg=Australia2[Australia2['Average'] == Australia2['Average'].min()]
#Calculating mean for unknown regions from 2000 to 2011
unknown1 = read_file.loc[read_file['Continent'] == 'unknown continent']
unknown2 = pd.concat([unknown1['Estimated HIV Prevalence% - (Ages 15-49)'],unknown1['Continent'],unknown1.loc[:,"2000":"2011"].mean(axis=1)],axis=1)
unknown2.columns = ['Region','Continent','Average']
unknown_highest_Avg=unknown2[unknown2['Average'] == unknown2['Average'].max()]
unknown_lowest_Avg=unknown2[unknown2['Average'] == unknown2['Average'].min()]
#Regions with highest average from each continent
Highest_Average = pd.concat([Asia_highest_Avg,Africa_highest_Avg,North_America_highest_Avg,South_America_highest_Avg,Europe_highest_Avg,Australia_highest_Avg,unknown_highest_Avg], ignore_index=True)
print("Highest Average HIV Estimated - Years 2000 to 2011\n")
print(Highest_Average)
Highest_Average.plot.bar(x="Continent", y="Average", title="Highest Average HIV Estimated - Years 2000 to 2011")
plt.show(block=True)
#Regions with lowest average from each continent
print("\nLowest Average HIV Estimated - Years 2000 to 2011\n")
Lowest_Average = pd.concat([Asia_lowest_Avg,Africa_lowest_Avg,North_America_lowest_Avg,South_America_lowest_Avg,Europe_highest_Avg,Australia_lowest_Avg,unknown_lowest_Avg], ignore_index=True)
Lowest_Average_drop_duplicates=Lowest_Average.drop_duplicates(subset=['Continent','Average'])
print(Lowest_Average_drop_duplicates)
Lowest_Average_drop_duplicates.plot.bar(x="Continent", y="Average", color="red",title="Lowest Average HIV Estimated - Years 2000 to 2011")
plt.show(block=True)
#overlaid bargraph
_, ax = plt.subplots()
Highest_Average.plot(kind='bar',x="Continent", y="Average" ,ax=ax, color='blue')
Lowest_Average_drop_duplicates.plot(kind='bar', x="Continent", y="Average", ax=ax, color='red')
```
